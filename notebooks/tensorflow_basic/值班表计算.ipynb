```python
# 人员类型列表
person_type1 = "person_type1"
person_type2 = "person_type2"
person_type3 = "person_type3"

# 岗位列表
position1 = "a"
position2 = "b"
position3 = "c"


person_type_dict = {}

for i in range(13):
    person_name = f"人员{i}"
    person_type_dict[person_name] = [person_type2]
for i in range(13, 26):
    person_name = f"人员{i}"
    person_type_dict[person_name] = [person_type1, person_type2]
for i in range(26, 35):
    person_name = f"人员{i}"    
    person_type_dict[person_name] = [person_type2, person_type3]
for i in range(35, 37):
    person_name = f"人员{i}"    
    person_type_dict[person_name] = [person_type1, person_type2, person_type3]   
position_type_dict = {
    position1: [person_type3],
    position2:[person_type1],
    position3:[person_type1, person_type2, person_type3]
}

total_person = len(person_type_dict)
person_count_dict = {}
for i in range(total_person): 
    person_name = f"人员{i}"    
    person_count_dict[person_name] = 0.0
```
